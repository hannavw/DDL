```python
import pandas as pd

#Put csv in dataframe with a seperation on ';'.
df1 = pd.read_csv('Assignment1stepsdata.csv', sep = ';')
df1.head()
```

```python
#Put csv in dataframe
df2 = pd.read_csv('Assignment1surveydata.csv')
df2.head()
```

```python
#Merge dataframes 
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
#Plot waist data.
df['waist'].plot()
```

```python
#Remove the values that are probably not correct
def remove_waist_values(x):
    if(x > 120.0): 
        return float('NaN')
    elif (x < 55.0):
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(remove_waist_values)
```

```python
#Plot a usable graph to analyse the data.
df['waist'].plot(kind = 'hist')
```
